```python
def convert(x):
    l = x.split(' - ')
    if len(l) == 2:
        return ((float(l[0]) + float(l[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
#import numpy as np
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for keys, subdf in df.groupby('location'):
        print(keys)
        m = np.mean(subdf.price_per_ft)
        s = np.std(subdf.price_per_ft)
        reduced_df = subdf[(subdf.price_per_ft > (m-s)) & (subdf.price_per_ft < (m+s))]
#         print(reduced_df)
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
#        print(df_out)
    return df_out
```
