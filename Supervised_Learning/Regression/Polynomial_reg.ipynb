```python
y = [ 100, 60, 50, 60, 60, 55, 60, 65, 70, 70, 75, 76, 78, 90, 99, 99, 100]
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
plt.scatter(X, y)
plt.show()
```

```python
y_predicted = np.poly1d(np.polyfit(X, y, 3))
y_predicted
```

```python
line = np.linspace(1, 17)
line
```

```python
model = y_predicted(line)
model
```

```python
plt.scatter(X, y)
plt.plot(line, model, color = 'r')
plt.show()
```

```python
x_predicted = y_predicted(X)
```

```python
x_predicted
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, x_predicted)
```

```python
coef = model.coef_
```

```python
intercept = model.intercept_
intercept
```
